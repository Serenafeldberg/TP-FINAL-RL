```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer logs
rewards = pd.read_csv('src/logs/rewards.csv')
losses = pd.read_csv('src/logs/losses.csv')

# Graficar rewards
plt.figure(figsize=(10, 5))
plt.plot(rewards['timestep'], rewards['reward'])
plt.xlabel('Timestep')
plt.ylabel('Episode Reward')
plt.title('Training Progress')
plt.show()

# Graficar losses
plt.figure(figsize=(10, 5))
plt.plot(losses['timestep'], losses['loss'])
plt.xlabel('Timestep')
plt.ylabel('Loss')
plt.title('Training Progress')
plt.show()
```
